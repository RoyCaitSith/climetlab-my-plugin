## Calculate track and intensity errors

```python
import datetime
import math
import numpy as np
import pandas as pd
import climetlab as cml
from tqdm import tqdm
from geopy.distance import great_circle

#cases = ['TROPICS_T_H1_ENS_C01', 'TROPICS_T_H1_ENS_C02', 'TROPICS_T_H1_ENS_C03']
cases = ['CON_TROPICS_T_H1_ENS_C01', 'CON_TROPICS_T_H1_ENS_C02', 'CON_TROPICS_T_H1_ENS_C03']

forecast_start_time = datetime.datetime(2021, 8, 26,  6, 0, 0)
initial_forecast_hours = 78
history_interval = 1
cycling_interval = 6
domain = 'd01'
dir_main = '/uufs/chpc.utah.edu/common/home/zpu-group16/cfeng/05_TROPICS/10_cycling_da_tropics_v3/track_intensity'
dir_best_track = '/uufs/chpc.utah.edu/common/home/u1237353/TROPICS/10_cycling_da_tropics_v3/track_intensity/best_track/V3'

for case in tqdm(cases):
    
    forecast_hours = initial_forecast_hours + cycling_interval*(int(case[-2:])-1)
    n_lead_time = int(forecast_hours/history_interval + 1)
    error_df = pd.DataFrame(0.0, index=np.arange(n_lead_time), columns=['Forecast_Hour', 'Track_Error (km)', 'MSLP_Error (hPa)', 'MWS_Error (Knot)'])

    filename = dir_main + '/' + case + '/multi/fort.69'
    df = pd.read_csv(filename, header=None, usecols=[2, 5, 6, 7, 8, 9])
    df.columns = ['Initial_Time', 'Forecast_Hour', 'Latitude', 'Longitude', 'MWS (Knot)', 'MSLP (hPa)']
    df.drop_duplicates(subset=['Forecast_Hour'], keep='last', inplace=True)

    Date_Time = []
    for it, fh in zip(df['Initial_Time'], df['Forecast_Hour']):
        Date_Time = Date_Time + [datetime.datetime.strptime(str(it), '%Y%m%d%H') + datetime.timedelta(hours = fh/100.0)]
    df.insert(loc=0, column='Date_Time', value=Date_Time)
    df.drop(columns=['Initial_Time', 'Forecast_Hour'], inplace=True)
    df.reset_index(drop=True, inplace=True)

    lat = list(df['Latitude'])
    lat = [x.split('N')[0] for x in lat]
    lat = pd.Series(map(float, lat))
    df['Latitude'] = 0.1*lat

    lon = list(df['Longitude'])
    lon = [x.split('W')[0] for x in lon]
    lon = pd.Series(map(float, lon))
    df['Longitude'] = -0.1*lon

    df.to_csv(dir_best_track + '/' + case + '_' + domain + '.csv', index=False)

    for idx, DT in enumerate(bt_df['Date_Time']):
        if datetime.datetime.strptime(DT, '%Y-%m-%d %H:%M:%S') == forecast_start_time: bt_str_index = idx
    for idx, DT in enumerate(df['Date_Time']):
        if DT == forecast_start_time: str_index = idx
        
    for idl in range(0, n_lead_time):

        idbt = math.floor(idl/6.0)
        df_index = str_index+idl
        bt_df_index = bt_str_index+idbt
        
        if df_index in df.index and bt_df_index in bt_df.index:
            index = idl
            error_df['Forecast_Hour'][index] = history_interval*idl + 6.0
            loc = (df['Latitude'][df_index], df['Longitude'][df_index])
            wbt = 1.0-(idl - 6.0*idbt)/6.0
            bt_loc = (wbt*bt_df['LAT'][bt_df_index] + (1.0-wbt)*bt_df['LAT'][bt_df_index+1],
                      wbt*bt_df['LON'][bt_df_index] + (1.0-wbt)*bt_df['LON'][bt_df_index+1])
            error_df['Track_Error (km)'][index] = great_circle(loc, bt_loc).kilometers
            error_df['MSLP_Error (hPa)'][index] = df['MSLP (hPa)'][df_index] - (wbt*bt_df['MSLP (hPa)'][bt_df_index] + (1.0-wbt)*bt_df['MSLP (hPa)'][bt_df_index+1])
            error_df['MWS_Error (Knot)'][index] = df['MWS (Knot)'][df_index] - (wbt*bt_df['MWS (Knot)'][bt_df_index] + (1.0-wbt)*bt_df['MWS (Knot)'][bt_df_index+1])

    error_df.to_csv(dir_best_track + '/Error_' + case + '_' + domain + '.csv', index=False)
```
